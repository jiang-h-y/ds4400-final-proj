```python
from tsfresh import extract_features, select_features
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
multivariate_gait_data = fetch_ucirepo(id=760)

# store data as df
df = multivariate_gait_data.data.features
```

```python
df
```

```python
ts_df = df.copy()
ts_df["subject_rep"] = df["subject"].astype("str") + "_" + df["replication"].astype("str")
ts_df["leg_joint"] = df["leg"].astype("str") + "_" + df["joint"].astype("str")
```

```python
ts_df
```

```python
extract_features(ts_df, column_id="subject_rep", column_kind="leg_joint", column_sort="time", column_value="angle")
```
